```python
#customdataset
class CustomDataset(Dataset):
  def __init__(self, df):
    self.df = df
    self.x = df.iloc[:,1:2].values
    self.y = df.iloc[:,2:].values
  def __len__(self):
    return len(self.x)
  def __getitem__(self, idx):
    x = tokenizer(self.x[idx][0])
    length = len(x['input_ids'])
    if length >= 64 :
      x['input_ids'] = torch.tensor(x['input_ids'][:64])
      x['attention_mask'] = torch.tensor(x['attention_mask'][:64])
    else :
      x['input_ids'] = torch.nn.functional.pad(torch.tensor(x['input_ids']), pad=(0, 64 - length), value=1.0)
      x['attention_mask'] = torch.nn.functional.pad(torch.tensor(x['attention_mask']), pad=(0, 64 - length), value=0.0)

    del x['token_type_ids']

    label = []
    label.append(label_ABS[idx])
    label.append(label_PN[idx])
    label.append(label_OFFEN[idx])
    
    return x, torch.tensor(label)
```
